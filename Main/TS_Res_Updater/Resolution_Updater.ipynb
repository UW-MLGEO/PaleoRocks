## Lower Image Resolution

```python
# Function to lower the resolution of an image
def lower_resolution(input_image_path, output_image_path, scale):
    # Open the image
   with Image.open(input_image_path) as img:
    # Calculate new dimensions
    new_width = int(img.width * scale)
    new_height = int(img.height * scale)
    
    # Resize the image
    img_resized = img.resize((new_width, new_height), Image.Resampling.NEAREST)
    
    # Save the resized image
    img_resized.save(output_image_path)
    print(f"Image saved at {output_image_path} with resolution {new_width}x{new_height}")

    # Call the function to lower the resolution
lower_resolution(input_image_path, output_image_path, scale)
```
