The goal of this code is to generate a point at the center of a thin section image patch to aid in image classification. 

```python
import numpy as np
import os
from skimage import io 
import cv2
from skimage import data
import matplotlib.pyplot as plt
```

```python
input_folder = "Fossil_Patches"
output_folder = "Fossil_Patches_CDot"

def centerdot(Image):
    width, height = Image.shape[:2]
    center_x = width//2
    center_y = height//2
    center_dot_img = cv2.circle(Image, (center_x, center_y), radius = 2, color = (65, 245, 200), thickness = -1)
    return center_dot_img

os.makedirs(output_folder, exist_ok=True)

# All files
for filename in os.listdir(input_folder):

    # Only process image files
    if filename.lower().endswith((".png", ".jpg", ".jpeg", ".tif")):
        
        filepath = os.path.join(input_folder, filename)
        image = io.imread(filepath)

        center_dot = centerdot(image)

        # Get base filename (without extension)
        base_name = os.path.splitext(filename)[0]

        # Save new image
        output_name = f"{base_name}_cdot.png"
        output_path = os.path.join(output_folder, output_name)
        io.imsave(output_path, center_dot)
```
